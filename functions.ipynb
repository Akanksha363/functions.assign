4. Write a Python function that checks if a given number is prime or not from 1 to 200.


```python
def is_prime(n):
    if n < 2:
        return False  # Numbers less than 2 are not prime
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False  # If divisible by any number in the range, it's not prime
    return True  # If not divisible by any number in the range, it's prime
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified
number of terms.



```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        # The line after the method definition needs to be indented
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


```python
def powers_of_two(exp):
   for i in range(exp + 1):
       yield 2 ** i
```

7. Implement a generator function that reads a file line by line and yields each line as a string.



```python
def read_file_line_by_line(file_path):
   with open(file_path, 'r') as file:
     for line in file:
        yield line.strip()
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.


```python
def remove_vowels(s):
 vowels = 'aeiouAEIOU'
 return ''.join(filter(lambda x: x not in vowels, s))
```

```python
string_with_vowels = "Hello, world!"
string_without_vowels = remove_vowels(string_with_vowels)
print(string_without_vowels)  # Output: Hll, wrld!
```
